```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
from scipy.stats import uniform, norm
import seaborn as sns
import scipy.linalg as LA
import sympy as sp
from time import perf_counter as timer
import random
%matplotlib
```

```python
L = 2024 # number of pieces of paper

outcomes = []
samplesize = 10000
for j in range(samplesize):
    
    Na = np.arange(1, L + 1) # my numbers on the paper pieces in an array
    Nl = Na.tolist() # my numbers on the paper pieces in a list
    
    x = random.randrange(0,L) # the value in this position, to start is equal to the number in it positions are 0 to 2023
    y = random.randrange(0,L)
    
    i = 0

    while i < (L - 1): # as repeats L - 1 times which is the number of iterations
            if x == y:
                y = random.randrange(0,L - i)
                #print(len(Nl))
            else:
                #num1 = Nl.pop(x) # as it considers the 1st element to be 0 stores value in position x as num1
                num1 = Nl[x]
                #print(num1)
                if x > y:
                    #num2 = Nl.pop(y) # as it considers the 1st element to be 0 stores value in position y as num2
                    num2 = Nl[y]
                else:
                    #num2 = Nl.pop(y - 1)
                    num2 = Nl[y - 1]
            z = abs(num1 - num2) 
            Nl.append(z)
            #print(len(Nl))
            Nl.pop(x)
            Nl.pop(y)
            x = random.randrange(0, L - 1 - i )
            y = random.randrange(0, L - 1 - i )
            i = i + 1       
    outcomes.append(Nl[0])
#print(len(Nl)) # - Testing
#print(Nl) # - Testing
#print(x,y,i) # - Testing
#print(outcomes) # - Testing    
```

```python
outcomes=np.array(outcomes)
mode = stats.mode(outcomes)
mean = np.mean(outcomes)
print("Modal output is:",mode[0]," with a count of", mode[1], " out of", samplesize, "iterations")
print("Mean output is:",mean)

plt.figure(figsize=(10, 6))
plt.hist(outcomes, bins=30, alpha=0.7, color='teal', label = 'Outcomes')
plt.xlabel('Outcome Values')
plt.ylabel('Frequency')
plt.axvline(mode[0], color='red', label = 'mode')
plt.axvline(mean, color='purple', label = 'mean')
plt.title('Outcome Distribution')
plt.legend()
plt.show()
```
