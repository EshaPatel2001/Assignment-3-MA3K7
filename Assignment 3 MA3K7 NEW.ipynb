```python
L = 2024 # number of pieces of paper

results = []
samplesize = 10000
for j in range(samplesize):
    
    Na = np.arange(1, L + 1) # my numbers on the paper pieces in an array
    Nl = Na.tolist() # my numbers on the paper pieces in a list
    
    x = random.randrange(0,L) # the value in this position, to start is equal to the number in it positions are 0 to 2023
    y = random.randrange(0,L)
    
    i = 0

    while i < (L - 1): # as repeats L - 1 times which is the number of iterations
            if x == y:
                y = random.randrange(0,L - i)
                #print(len(Nl))
            else:
                #num1 = Nl.pop(x) # as it considers the 1st element to be 0 stores value in position x as num1
                num1 = Nl[x]
                #print(num1)
                if x > y:
                    #num2 = Nl.pop(y) # as it considers the 1st element to be 0 stores value in position y as num2
                    num2 = Nl[y]
                else:
                    #num2 = Nl.pop(y - 1)
                    num2 = Nl[y - 1]
            z = abs(num1 - num2) 
            Nl.append(z)
            #print(len(Nl))
            Nl.pop(x)
            Nl.pop(y)
            x = random.randrange(0, L - 1 - i )
            y = random.randrange(0, L - 1 - i )
            i = i + 1       
    results.append(Nl[0])
#print(len(Nl)) # - Testing
#print(Nl) # - Testing
#print(x,y,i) # - Testing
#print(v) # - Testing    
```

```python
results=np.array(results)
mode = stats.mode(results)
mean = np.mean(results)
print("Modal output is:",mode[0]," with a count of", mode[1], " out of", samplesize, "iterations")
print("Mean output is:",mean)

plt.figure(figsize=(10, 6))
plt.hist(results, bins=30, alpha=0.7, color='teal', label = 'results')
plt.xlabel('results Values')
plt.ylabel('Frequency')
plt.axvline(mode[0], color='red', label = 'mode')
plt.axvline(mean, color='purple', label = 'mean')
plt.title('results Distribution')
plt.legend()
plt.show()
```

```python
#NEW NEW CODE DEFINETLY WORKS CORRECTLY THROUGH VARIOUS TESTING

L = 2024

samplesize = 10


def hatgame(number_list):
    while len(number_list) > 1:
        # Choose two random numbers from the array
        x = random.randint(0, len(number_list) - 1) # the x-th position in our number list
        y = random.randint(0, len(number_list) - 1) # the x-th position in our number list

        # Ensure x and y are distinct
        while x == y:
            y = random.randint(0, len(number_list) - 1)

        number1 = number_list[x]
        number2 = number_list[y]

        # Calculate the absolute difference
        abs_diff = abs(number1 - number2)

        # Add the absolute difference to the array
        number_list.append(abs_diff)

        # Remove the two picked numbers from the array
        number_list.pop(max(x, y))
        number_list.pop(min(x, y))

    return number_list[0] # returns that last number in the hat as a number not array

results = []

for i in range(samplesize):
    pieces_of_paper = list(range(1, L + 1))
    result = hatgame(pieces_of_paper)
    results.append(result)
    
#print(results) - Testing

#making histogram with results to show frequency of the results
#results=np.array(results)
mode = stats.mode(results)
mean = np.mean(results)
print("Modal output is:",mode[0]," with a count of", mode[1], " out of", samplesize, "iterations")
print("Mean output is:",mean)

plt.figure(figsize=(10, 6))
plt.hist(results, bins=30, alpha=0.7, color='teal', label = 'results of game')
plt.xlabel('result Values')
plt.ylabel('Frequency')
plt.axvline(mode[0], color='red', label = 'mode')
plt.axvline(mean, color='purple', label = 'mean')
plt.title('Result Distribution')
plt.legend()
plt.show()
```
